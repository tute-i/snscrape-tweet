```python
import pandas as pd
import snscrape.modules.twitter as sntwit
```

```python
max_tweet = 50
tweet_list = []
for t, tweet in enumerate(sntwit.TwitterSearchScraper('from : elonmusk').get_items()):
 if t > max_tweet:
      break
 else:
          tweet_list.append([tweet.date, tweet.id, tweet.content,tweet.user.username,tweet.hashtags,tweet.cashtags])
```

```python
tweet_list[2]
```

```python
elonmusk_df = pd.DataFrame(tweet_list, columns = ["tarih", "tweet ID", "içerik", "kullanıcı","hashtags","cashtags"])
elonmusk_df.head(10)
```
